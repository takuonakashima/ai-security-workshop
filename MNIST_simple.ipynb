```python
# 1. 必要なライブラリの読み込み
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# 2. データをAIが処理しやすい形（テンソル）に変換する設定
transform = transforms.Compose([
    transforms.ToTensor()
])

# 3. MNISTデータセットのダウンロード（インターネットからColab上に保存）
print("MNISTデータセットをダウンロード中...")
train_dataset = torchvision.datasets.MNIST(root='./data',
                                           train=True,
                                           download=True,
                                           transform=transform)

# 4. データを1枚ずつ取り出すための準備（データローダー）
train_loader = torch.utils.data.DataLoader(train_dataset,
                                           batch_size=1,
                                           shuffle=True)

# 5. データの中身を取り出して、1枚だけ画像として表示する
# イテレータを使って最初のデータを取得
dataiter = iter(train_loader)
images, labels = next(dataiter)

# 画像の表示設定
image_np = images[0].numpy().squeeze() # PyTorch特有の配列の形を画像用に整える
plt.imshow(image_np, cmap='gray')
plt.title(f"正解ラベル（AIに当てさせる答え）: {labels[0].item()}")
plt.axis('off') # グラフの軸を隠す
plt.show()
```
